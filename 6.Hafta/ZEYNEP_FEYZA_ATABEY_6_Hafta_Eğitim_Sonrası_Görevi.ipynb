Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1. problemimiz için kodlarını buraya yazabilirsin.

import random
import time

class Kumanda():

	def __init__(self, durum ="KAPALI", ses_duzeyi = 0, kanal_listesi = ["Disney","National Geographic"], kanal = "Disney"):
		self.durum = durum
		self.ses_duzeyi = ses_duzeyi
		self.kanal_listesi = kanal_listesi
		self.kanal = kanal

	def open(self):
		if(self.durum == "AÇIK"):
			print("Televizyon açık.")
		else:
			self.durum = "KAPALI"
			print("Televizyon açılacak.")

	def close(self):
		if(self.durum == "KAPALI"):
			print("Televizyon kapalı.")
		else:
			self.durum = "KAPALI"
			print("Televizyon kapanacak.")

	def ses_ayarı(self):
		while(True):
			secim = input("Ses Düzeyi - '<'\nSes Seviyesi + '>'\nÇıkış 'X'\nSeç: ")

			if(secim == "X"):
				break

			if(secim == "<"):
				if(self.ses_duzeyi <= 0):
					print("Televizyonun sesini daha fazla kısamazsınız.")
				else:
					self.ses_duzeyi -= 1
					print("Televizyonun şu anki ses düzeyi: ",self.ses_duzeyi)

			elif(secim == ">"):
				if(self.ses_duzeyi >= 100):
					print("Televizyonun sesini daha fazla açamazsınız.")
				else:
					self.ses_duzeyi += 1
					print("Şu anki ses seviyesi: ",self.ses_duzeyi)

			else:
				print("Bilinmeyen komut")

			print("\n----------------\n")

	def ekle_kanal(self, name):
		self.kanal_listesi.append(name)
		time.sleep(1)
		print("Yeni kanal eklendi.")

	def rasgele_degistir(self):
		random_number = random.randint(0, len(self.kanal_listesi) - 1)
		self.kanal = self.kanal_listesi[random_number]
		print("Kanal değişti.\nŞu anki kanal ismi : ",self.kanal)

	def __len__(self):
		return len(self.kanal_listesi)

	def __str__(self):
		return "---Televiyon Hakkında Bilgileri---\nTV'nin Şu Anki Durumu: {}\nTV'nin Şu Anki Ses Düzeyi: {}\nTV Kanalı Listesi: {}\nTV Şu Anki Kanal: {}".format(self.durum, self.ses_duzeyi, self.kanal_listesi, self.kanal)
	


def main():

	tv = Kumanda()

	print("""
		-----------Kumanda Komutları-----------
		1. TV Aç
		2. Tv Kapat
		3. Ses Ayarları
		4. Kanal Ekle
		5. Kanal Sayısını Öğrenme
		6. Rasgele Kanala Geçme
		7. Televizyon Bilgileri
		q. Çıkış
		""")

	while(True):

		komut = input("Lütfen kumandaya basarak komut giriniz: ")

		if(komut == "q"):

			print("Kumanda ve komut sistemi kapanıyor.")
			time.sleep(1)
			break

		elif(komut == "1"):
			tv.open()

		elif(komut == "2"):
			tv.close()

		elif(komut == "3"):
			tv.ses_ayarı()

		elif(komut == "4"):
			name = input("Lütfen kanal ismi giriniz: ")
			tv.ekle_kanal(name)

		elif(komut == "5"):
			print("Kanalların sayısı: ", len(tv))

		elif(komut == "6"):
			tv.rasgele_degistir()

		elif(komut == "7"):
			print(tv)

		else:
			print("Uyarı! Bilinmeyen Komut")


if __name__ == "__main__":
	main()	
```
