# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1. problemimiz için kodlarını buraya yazabilirsin.

import time

print("-------------------------------------------------------")
print("|                                                      |")
text =" -- YetGen VKİ Veri Giriş Platformuna Hoş Geldiniz! --"
x = text.center(20,"*")
print(x)
print("|                                                      |")
print("-------------------------------------------------------")

def VKİ(boy, kilo):
    vki = kilo/(boy**2)
    return round(vki,1)
  
def veri_yorumlama(vki):
  if (vki < 18.5):
      print("Zayıf")
 
  elif ( vki >= 18.5 and vki < 24.9):
    print("Normal Kilolu")
 
  elif ( vki >= 24.9 and vki < 30):
    print("Fazla Kilolu")
 
  elif ( vki >=30):
    print("Obez")

  elif (vki >= 40):

    print("İleri derece obez (morbide obez)")

def create_output(isim_soyisim,yas,boy,kilo,vki,yorum):
  
  record_time = time.ctime(time.time())
  giris1 = f"Kullanıcının adı: {isim_soyisim}\n"
  giris2 = f"Kullanıcın yaşı: {yas}\n "
  giris3 = f"Kullanıcın boyu: {boy}\n" 
  giris4 = f"Kullanıcın kilosu: {kilo}\n" 
  giris5 = f"Kullanıcın vücut-kitle-indexi : {vki}\n"
  giris6 = f"Girilen değer aralığı {yorum}\n" 
  giris7 = f"Kaydın oluşturulma tarihi: {record_time}\n"

  cıktı = giris1 + giris2 + giris3 + giris4 + giris5 + giris6 + giris7
  return cıktı

def main():
  input_number = int(input("Kaç kullanıcının verisini girmek istiyorsunuz? "))
  for info in range(input_number):
    print(f"Kullanıcın kaydı {info+1}")
    isim_soyisim = input("Kullanıcın isim-soyismini giriniz: ")
    yas = int(input("Kullanıcın yaşını giriniz: "))
    kilo = float(input("Kullanıcın kilosunu (kg) giriniz: "))
    boy = float(input("Kullanıcın boyunu (m) giriniz: "))
    vki = VKİ(boy,kilo)
    sonuc = veri_yorumlama(vki)

    cıktı = create_output(isim_soyisim,yas,boy,kilo,vki,sonuc)

    with open("veriler.txt","a") as f:
      f.write(cıktı)
      f.write("-"*20)
      f.write("\n")
      print("\nKullanıcı verileri kaydedildi.")

  print("---------------------------------------------------------")
  print("YetGen VKİ Veri Sistemini Kullandığınız İçin Teşekkürler!\nBütün veriler veriler.txt dosyasına kaydedilmiştir.\n")

main()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2. problemimiz için kodlarını buraya yazabilirsin.

menu = {
    "Elma": 3,
    "Ahududu" : 4
    ,"Böğürtlen": 4.5,
    "Şeftali": 3.5 , 
    "Muz": 2.5,
    "Karpuz": 5 ,
    "Kavun": 5.5
      }

# print("Elmanın fiyatı:", menu["Elma"], "₺")

def menu_goruntuleme():
  print("-----YetBar-Smoothie-Menü-------")

  for item in list(menu.keys()):
      print(f"{item.upper():<10} : {menu[item]} ₺")

def fiyat_hesaplama(meyve,sayı):
  assert sayı >= 0 , "Seçim sayınız 0 ve 0'dan küçük olmamalıdır."

  if meyve not in list(menu.keys()):
    return None
  else:
    return menu[meyve] * sayı

def smoothie_secim():
  menu_goruntuleme()
  print("Smoothie'nizde bulunmasını istediklerinizi seçiniz. Çıkış yapmak için '0'ı tuşlayınız.")

  secilen_meyveler = []
  secim_sayısı = []

  odenecek_fiyat = 0

  while 1:
    secilen_urun = input("Smoothie'nizi yapmak istediğiniz meyveyi giriniz: ")
    if secilen_urun == "0":
      print("Fiş kaydedilip ödenecek tutar hesaplanıyor...")
      break
    eklenecek_urun_sayısı = int(input(f"Smoothie'nizinde kaç tane {secilen_urun} bulunsun istiyorsunuz? "))
    fiyat = fiyat_hesaplama(secilen_urun,eklenecek_urun_sayısı)
    if fiyat is None:
      print("Meyve girildi, yeni meyve giriniz..")
    else:
      print(f"Ödenecek tutar: {fiyat} ₺ .\n Onaylamak için 'evet' iptal etmek için 'hayır' ifadesini giriniz.")
      kontrol = input("Satın almayı onaylıyor musunuz? ")
      if kontrol == "evet":
        print("Onaylandı.")
        print("-"*40)
        odenecek_fiyat += fiyat
        secilen_meyveler.append(secilen_urun)
        secim_sayısı.append(eklenecek_urun_sayısı)
        print(f"Ara Toplam: {odenecek_fiyat} ₺")
      elif kontrol == "hayır":
        print("Satın alma iptal edildi.")
        print("-"*40)

  return odenecek_fiyat , secilen_meyveler , secim_sayısı

def main():
  print("-----Sağlıklı Smoothie Yapabileceğiniz YetBar'a Hoş Geldiniz!-------\nMenüyü görmek için '1'e, Smoothie'nizi yapmaya başlamak için '2'ye basınız.\n Otomattan çıkmak için : '0'a basınız.")
  akıs = None
  while 1:
    print("--------------------------------------------------------------------")
    akıs = int(input("Tercihinizi giriniz: "))
    if akıs <= 0 or akıs > 3:
      print("Çıkış yapılıyor...\nGüzel günler dileriz.")
      break
    elif akıs == 1:
      print("------------------------------------------------------------------")
      menu_goruntuleme()
    elif akıs == 2:
      odenecek_fiyat, secilen_meyveler , secim_sayısı = smoothie_secim()

      with open("fis.txt","a") as f:
        f.write(str(time.ctime(time.time())))
        f.write("\n")
        f.write(",".join(map(str,secilen_meyveler)))
        f.write("\n")
        f.write(f"Toplam: {odenecek_fiyat} ₺")
        f.write("\n")
        f.write("*"*20)
        f.write("\n")

      cıkıs = input("Başka bir işlem yapacak mısın? (Evet-Hayır):")

      if cıkıs == "Evet":
        pass
      else:
        print("-------------------------------------------------------")
        print("|                                                      |")
        print(f"     Ödemeniz gereken tutar {odenecek_fiyat} ₺'dir.")
        print("|                                                      |")
        print("\n  Sağlıklı güzel günler dileriz.\n   Bizi Tercih Ettiğiniz İçin Teşekkürler!\n  ----------------Afiyet Olsun!-------------")
        print("|                                                      |")
        print("-------------------------------------------------------")
        break

main()
```
